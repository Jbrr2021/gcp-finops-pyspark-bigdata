```python
# =================================================================================
# ENGENHARIA DE DADOS EM BIG DATA / PROPRIEDADES DE DATA_FRAMES DISTRIBUÍDOS
# CONCEITO: Manipulação e Criação de Colunas usando com .withColumn(), col() e round()
#
# CONTEXTO DE NEGÓCIO (Pedido do Gestor):
# "O nosso Data Lake de faturamento recebeu uma volumetria massiva de logs.
# Crie um pipeline distribuído com PySpark que ingira esses registros e gere uma
# nova coluna chamada 'custo_com_imposto', aplicando uma taxa tributária internacional
# de 15% sobre o custo bruto original e arredondando os valores com duas casas decimais."
# =================================================================================

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, round

# Initialize SparkSession
spark = SparkSession.builder.appName("DataTransformation").getOrCreate()

# 1. Criando os dados brutos massivos simulados (Lista de Tuplas)
dados_faturamento = [
    ("ipnet-vendas-prod", "Compute Engine", 450.00),
    ("ipnet-data-lake", "BigQuery", 320.50),
    ("ipnet-vendas-prod", "Cloud Storage", 25.00)
]

# Definindo o nome das colunas correspondentes
colunas_tabela = ["projeto_id", "servico_descricao", "custo_bruto"]

print("⚡ Ingerindo dados na memória distribuída do Spark...")
# 2. Criando o DataFrame oficial do PySpark
df_spark = spark.createDataFrame(dados_faturamento, colunas_tabela)

print("🛠️ Aplicando transformação de Big Data (Taxa de Imposto)...")
# 3. O SEU DESAFIO: Use o .withColumn para criar a coluna 'custo_com_imposto'
# A fórmula matemática é: col("custo_bruto") * 1.15
# Use a função round(..., 2) por fora para arredondar os centavos!
df_transformado = df_spark.withColumn(
    "custo_com_imposto",
    round(col("custo_bruto") * 1.15, 2)
)

print("\n📊 --- VISUALIZAÇÃO DATAFRAME DISTRIBUÍDO PYSPARK ---")
# 4. Use o comando oficial do Spark para exibir o resultado final na tela
df_transformado.show()
print("-----------------------------------------------------")
```

```python
# =================================================================================
# ENGENHARIA DE DADOS EM BIG DATA / FILTRAGEM EM CLUSTERS DISTRIBUÍDOS
# CONCEITO: Otimização de Busca e Filtragem de Linhas com .filter() e col()
#
# CONTEXTO DE NEGÓCIO (Pedido do Gestor):
# "A diretoria de compliance precisa analisar apenas os gastos de alta criticidade.
# Aplique um filtro na nossa tabela distribuída do Spark para isolar e exibir na tela
# APENAS os recursos cujo custo_bruto original seja maior do que R$ 100,00."
# =================================================================================

print("🔍 Filtrando registros críticos na memória distribuída do Spark...")

# 1. Aplicando o filtro distribuído (Equivalente ao WHERE do SQL)
df_critico = df_transformado.filter(col("custo_bruto") > 100.00)

print("\n📊 --- RELATÓRIO COMPLIANCE: CUSTOS ACIMA DE R$ 100,00 ---")
# 2. Exibindo as linhas filtradas na tela
df_critico.show()
print("----------------------------------------------------------")
```
